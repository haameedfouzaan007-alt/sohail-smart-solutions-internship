## Part 1 — Learn

### Why AI Work Needs a Data Tool

AI work needs data. At first, we can store data using lists, dictionaries, or objects. But when the data becomes bigger, it becomes difficult to manage everything manually.

For example, if we have only 3 students, using objects is fine. But if we have hundreds or thousands of records, it becomes harder to calculate averages, filter data, or count values using only normal Python loops.

This is why pandas is useful. Pandas helps us work with data in table format. It makes it easier to load, inspect, filter, and analyze data.

### DataFrame and Series

A DataFrame is like a table in pandas. It has rows and columns, similar to an Excel sheet.

A Series is one column from a DataFrame. For example, if the full table is a DataFrame, then the `grade` column alone is a Series.

### Loading Data from CSV

A CSV file stores data in table format. In pandas, we can load a CSV file using:

`pd.read_csv("students.csv")`

This reads the CSV file and converts it into a DataFrame.

### Inspecting Data

Pandas has useful commands to understand the data:

- `.head()` shows the first few rows.
- `.info()` shows column names and data types.
- `.describe()` gives basic statistics.
- `.shape` shows the number of rows and columns.

### Selecting Columns and Filtering Rows

We can select one column like this:

`df["grade"]`

We can filter rows like this:

`df[df["grade"] > 80]`

This shows only the students who scored above 80.

### Basic Statistics

Pandas can calculate basic statistics easily.

For example:

`df["grade"].mean()`

This gives the average grade.

Also:

`df["course"].value_counts()`

This counts how many times each course appears.

```python
import pandas as pd

data = {
    "student_id": ["S001", "S002", "S003"],
    "name": ["Haameed", "Ahmed", "Sara"],
    "course": ["Python OOP Level 2", "Python OOP Level 2", "Python OOP Level 2"],
    "grade": [85, 78, 92]
}

df = pd.DataFrame(data)

df.to_csv("students.csv", index=False)

print("students.csv created successfully.")
```

## Short Note — OOP Classes vs pandas

In my OOP project, I used classes like `Student` and `Course` to organize the data. This was useful for building a small program because each object had its own details and methods.

With pandas, the data is handled more like a table. This makes it easier to inspect, filter, and calculate from the data.

For example, in OOP I may need to write loops to calculate the average grade. In pandas, I can do it in one line using:

`df["grade"].mean()`

So, OOP is useful for designing programs and objects. Pandas is useful for working with datasets and doing analysis.
