# Week 5 — Data Cleaning with pandas

In this notebook, I am practicing data cleaning using pandas.  
For this task, I used a deliberately messy student dataset based on my Week 4 student analytics data.

The dataset includes common real-world data problems such as missing values, duplicate rows, inconsistent text formatting, and wrong data types.

```python
# Creating a deliberately messy CSV file for the task

messy_data = """student_id,name,university,course,grade
S001, haameed , AURAK ,python oop,85
S002,AHMED,aurak,Python OOP,78
S003,Sara,Aurak ,DATA HANDLING,92
S003,Sara,Aurak ,DATA HANDLING,92
S004,,AURAK,data handling,88
S005,Omar, aurak ,AI Basics,seventy four
S006,Mariam,, ai basics ,90
S007,Yousef,AURAK,Python OOP,
S008, lina,aurak ,python oop,82
"""

with open("messy_students.csv", "w") as file:
    file.write(messy_data)

print("messy_students.csv file created successfully.")
```

```python
import pandas as pd

df = pd.read_csv("messy_students.csv")

print("Original Messy Dataset:")
df
```

## Original Messy Dataset

The dataset has been intentionally made messy for practice.  
Some of the problems include:

- Extra spaces in names and university values
- Different capitalization in text columns
- One duplicate student row
- One missing student name
- One missing university value
- One missing grade
- One grade written as text instead of a number

```python
print("Dataset Information:")
df.info()
```

```python
print("Missing Values in Each Column:")
print(df.isnull().sum())
```

```python
print("Number of Duplicate Rows:")
print(df.duplicated().sum())
```

```python
print("Data Types:")
print(df.dtypes)
```

```python
# Creating a copy of the original dataset before cleaning
df_cleaned = df.copy()
```

```python
# Fixing the wrong text value in the grade column
df_cleaned["grade"] = df_cleaned["grade"].replace("seventy four", "74")

# Converting grade column into numeric data type
df_cleaned["grade"] = pd.to_numeric(df_cleaned["grade"], errors="coerce")

print("Grade column after converting to numeric:")
print(df_cleaned["grade"])
```

```python
# Handling missing values

# Remove rows where student name is missing because name is important
df_cleaned = df_cleaned.dropna(subset=["name"])

# Fill missing university values with "Unknown"
df_cleaned["university"] = df_cleaned["university"].fillna("Unknown")

# Fill missing grade values with the average grade
average_grade = df_cleaned["grade"].mean()
df_cleaned["grade"] = df_cleaned["grade"].fillna(average_grade)

# Round grade values to 2 decimal places
df_cleaned["grade"] = df_cleaned["grade"].round(2)

print("Missing values after handling:")
print(df_cleaned.isnull().sum())
```

### Explanation for Handling Missing Values

For the missing university value, I filled it with `"Unknown"` because university is a text column. It does not directly affect grade calculations, so keeping the student record is still okay.

For the missing grade value, I filled it with the average grade because grade is a numeric column and it is needed for analysis. If I filled the grade with `"Unknown"`, pandas would treat the grade column as text, and calculations like average grade would not work properly.

So I handled missing values differently depending on the type and importance of the column.

```python
# Removing duplicate rows
df_cleaned = df_cleaned.drop_duplicates()

# Resetting the index after removing rows
df_cleaned = df_cleaned.reset_index(drop=True)

print("Number of duplicate rows after cleaning:")
print(df_cleaned.duplicated().sum())
```

```python
# Standardizing name column
df_cleaned["name"] = df_cleaned["name"].str.strip().str.title()

# Standardizing university column
df_cleaned["university"] = df_cleaned["university"].str.strip().str.lower()
df_cleaned["university"] = df_cleaned["university"].replace({
    "aurak": "AURAK",
    "unknown": "Unknown"
})

# Standardizing course column
df_cleaned["course"] = df_cleaned["course"].str.strip().str.lower()
df_cleaned["course"] = df_cleaned["course"].replace({
    "python oop": "Python OOP",
    "data handling": "Data Handling",
    "ai basics": "AI Basics"
})

print("Text columns after standardization:")
df_cleaned
```

```python
print("Cleaned Dataset Information:")
df_cleaned.info()
```

```python
print("Missing Values After Cleaning:")
print(df_cleaned.isnull().sum())
```

```python
print("Duplicate Rows After Cleaning:")
print(df_cleaned.duplicated().sum())
```

```python
print("Data Types After Cleaning:")
print(df_cleaned.dtypes)
```

```python
print("Before Cleaning:")
display(df)

print("After Cleaning:")
display(df_cleaned)
```

```python
df_cleaned.to_csv("cleaned_students.csv", index=False)

print("cleaned_students.csv file saved successfully.")
```

## Summary: The 3 Cleaning Steps That Mattered Most and Why

The first important cleaning step was handling missing values. One student name was missing, so I removed that row because the student could not be identified properly. For the missing university value, I filled it with `"Unknown"` because it was not the main column used for grade analysis. For the missing grade value, I filled it with the average grade so that the student record could still be used.

The second important step was removing duplicate rows. One student record was repeated, and keeping it would have affected the student count and average grade. Removing duplicates made the dataset more accurate.

The third important step was standardizing text and fixing the grade data type. Some names, university values, and course names had extra spaces or different capitalization. I cleaned them using string methods like `.str.strip()` and `.str.lower()`. I also converted the grade column into numeric form because grades must be numbers for calculations such as averages.

Overall, this task helped me understand that data cleaning is an important step before doing analysis or machine learning. Even a small dataset can have problems, and if those problems are not fixed, the final results can become wrong.
